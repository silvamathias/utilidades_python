```python
sub_lista_qnt_fx = lambda lista, div: [lista[i::div] for i in range(div)] #Mesmo resultado que 'sublista_qnt_fixa' mas com método diferente

def sublista_qnt_fixa(qnt_sublista):
    '''
    Decorador:
    divide a lista informada pela quantidades de vezes solicitadas
    '''
    def parametro(lista_ex):
        def parte_lista(lista_in):
            list = lista_ex(lista_in)
            n = qnt_sublista
            return [list[i::n] for i in range(n)]
        
        return parte_lista
    return parametro

def sublista_item_fixo(qnt_itens):
    '''
    Decorador:
    divide a lista informada quantas vezes for necessárias para que cada sublista tenha a quantidade de itens tenda para o valor informado
    '''
    def parametro(lista_ex):
        def parte_lista(lista_in):
            list = lista_ex(lista_in)
            n = int((len(list) / qnt_itens) + 1)
            return [list[i::n] for i in range(n)]
        
        return parte_lista
    return parametro

def retira_caractere(lista):
    '''
    Decorador
    Retira os caracteres especiais que não são removidos com 'normaliza_nome'
    '''
    def formata_txt(item):
        txt = lista(item)
        return [item.replace(' (r$/h)','').replace(' ','_').replace('.','').replace('/','_') for item in txt]
        
    
    #return [formata_txt(item) for item in lista]
    return formata_txt

@retira_caractere
def normaliza_nome(str):
    '''
    Retira acentuação das strings e converte tudo para letra minúscula
    '''
    def limpa_txt(txt):
        texto = normalize('NFKD', txt)
        return texto.encode('ASCII', 'ignore').decode('ASCII')

    return [limpa_txt(txt).lower() for txt in str]

@sublista_item_fixo(990)
def concatena_df(df):
    'Retorna uma lista onde cada item é a linha concatenada, separada por ',', do DF informado'
    lin_lista = []
    lin_n = ''
    for lin in df.iterrows():
        #print(lin[1])
        for col in lin[1]:
            #print(col)
            if lin_n == '':
                lin_n = "'" + str(col).replace("'", "'+ char(39) +'") + "'"
            else:
                lin_n += ",'" + str(col).replace("'", "'+ char(39) +'") + "'"
        
        lin_lista += [lin_n]
        lin_n = ''
    
    return lin_lista

def concatenar_lista(lista):
    '''
    Transforma uma lista em uma string separada por ','
    '''
    txt = ''
    for item in lista:
        if txt == '':
            txt = item
        else:
            txt += ',' + item

    return txt

```
